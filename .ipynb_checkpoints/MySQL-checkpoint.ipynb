```python
import sqlalchemy
import pymysql
from sqlalchemy import create_engine
import pandas as pd
import os
```

```python
password = 'TigerW00ds'
db_name = 'library'
```

```python
books = pd.read_csv(os.getcwd() + '\\Books.csv')
authors = pd.read_csv(os.getcwd() + '\\Authors.csv')
patrons = pd.read_csv(os.getcwd() + '\\Patrons.csv')
loans = pd.read_csv(os.getcwd() + '\\Loans.csv')
```

```python
conn = pymysql.connect(host='127.0.0.1', port=3306, user='root', password=password, database=db_name)
```

```python
connection = create_engine("mysql+pymysql://root:" + password + "@localhost:3306/" + db_name)
```

```python
books
table_name = 'books'
books.to_sql(name=table_name, con=connection, if_exists='replace', index=False)
```

```python
authors
table_name = 'authors'
authors.to_sql(name=table_name, con=connection, if_exists='replace', index=False)
```

```python
patrons
table_name = 'patrons'
patrons.to_sql(name=table_name, con=connection, if_exists='replace', index=False)
```

```python
loans 
table_name = 'loans'
loans.to_sql(name=table_name, con=connection, if_exists='replace', index=False)
```

```python
loans["New_Fine"] = loans.loc[:,"Fine"]
loans["New_Fine"] = loans["New_Fine"].map(lambda x: x.lstrip("$"))
loans["New_Fine"] = loans["New_Fine"].astype("float")
loans = loans.drop('Fine', axis = 1)
loans.to_sql(name=table_name, con=connection, if_exists='replace', index=False)
```

```python
query1 = "SELECT * FROM loans"
result_df = pd.read_sql(query1, con = connection)
result_df
```

```python
query2 = """
SELECT `Patron First Name`, `Patron Last Name`, New_Fine FROM loans
Order by New_Fine;
"""
```

```python
result_df = pd.read_sql(query2, con=connection)
result_df
```

```python
query3 = """
SELECT MAX(New_Fine) AS Max_fine
FROM loans;
"""
```

```python
result_df = pd.read_sql(query3, con=connection)
result_df
```

```python
query3 = """
SELECT AVG(New_Fine) AS Avg_fine
FROM loans
WHERE Return_Date > Due_Date;
"""
result_df = pd.read_sql(query3, con = connection)
result_df
```

```python
query4 = """
SELECT * from loans
WHERE New_Fine > 0;
"""
result_df = pd.read_sql(query4, con = connection)
result_df
```

```python
query5 = """
SELECT Title, COUNT(Title) AS count
FROM Books
GROUP BY Title
ORDER BY count DESC;
"""
result_df = pd.read_sql(query5, con = connection)
result_df
```
